### Problem 1 Student Performance

```python
# Sorting out the dataframe for grades
student_performance = pd.read_csv(r"C:\Users\rgs88\OneDrive\Documents\archive (21)\StudentsPerformance.csv")
student_performance
```

```python
# Mean across all grades
student_performance["score"] = student_performance[["math score", "reading score", "writing score"]].mean(axis = 1)
```

```python
# Assigning grades function, where we place 60 as Fail, 60 to 80 as B, 80+ as A
def assign_grade(score):
    if score < 60:
        return "Fail"
    elif score >= 60 and score <= 80:
        return "B"
    else:
        return "A"

student_performance["grade"] = student_performance["score"].apply(assign_grade)
```

```python
student_performance
```

```python
# Set a random seed for reproducibility
np.random.seed(0)

# Create an instance of the Faker class for generating random names
fake = Faker()

# Create a list of random student names
students = [fake.name() for _ in range(10)]

# Create a dictionary to store the data
data = {
    'Student_Name': students,
    'Biology': np.random.randint(0, 51, 10),  # Random marks out of 50 for Biology
    'Chemistry': np.random.randint(0, 51, 10),  # Random marks out of 50 for Chemistry
    'Maths': np.random.randint(0, 51, 10),  # Random marks out of 50 for Maths
    'Physics': np.random.randint(0, 51, 10),  # Random marks out of 50 for Physics
    'Computer_Science': np.random.randint(0, 51, 10),  # Random marks out of 50 for Computer Science
}

# Create a DataFrame
student_df = pd.DataFrame(data)
```

```python
student_df
```

```python
# Calculate percentages and create new columns
subject_columns = ['Biology', 'Chemistry', 'Maths', 'Physics', 'Computer_Science']

for subject in subject_columns:
    percent_column = subject + '_Percent'
    student_df[percent_column] = (student_df[subject] / 50) * 100
```

```python
student_df
```

```python
def assign_grade(score):
    if score < 60:
        return "Fail"
    elif score >= 60 and score <= 80:
        return "B"
    else:
        return "A"

student_df["Biology_Score"] = student_df["Biology_Percent"].apply(assign_grade)
student_df["Chemistry_Score"] = student_df["Chemistry_Percent"].apply(assign_grade)
student_df["Maths_Score"] = student_df["Maths_Percent"].apply(assign_grade)
student_df["Physics_Score"] = student_df["Physics_Percent"].apply(assign_grade)
student_df["Computer_Science_Score"] = student_df["Computer_Science_Percent"].apply(assign_grade)
```
